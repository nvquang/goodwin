```python
import requests
import json

for tit in arr[3559:]:
    string = 'https://goodwin.api.relinklabs.com/v1/explore?title='
    string+= tit.replace(' ','%20')
    r = requests.get(string)
    if(r.status_code == 200):
        with open('result.txt', 'a') as outfile:
            json.dump(r.json(),outfile)
            outfile.write('\n')
    else:
        with open('error_message.txt', 'a') as messfile:
            json.dump(r.json(),messfile)
            messfile.write('\n')
            
        with open('error_title.txt', 'a') as titfile:
            titfile.write(tit+','+ str(r.status_code))
            titfile.write('\n')  
```
